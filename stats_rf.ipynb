```python
import csv

razvodi = dict()
with open('razvodi.stat', 'r', encoding = 'utf8') as file:
    lines = csv.reader(file)
    
    for i in lines:
        if i[0] == 'Годы':
            continue
        else:
            razvodi[int(i[0])] = int(i[1])

print(razvodi)
```

```python
import csv

braki = dict()

with open('braki.stat', 'r', encoding = 'utf8') as file:
    lines = csv.reader(file)

    for i in lines:
        if i[0] == 'Годы':
            continue
        else:
            braki[int(i[0])] = int(i[1])
```

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

x_coords1 = list(braki.keys())
y_coords1 = list(braki.values())

x_coords2 = list(razvodi.keys())
y_coords2 = list(razvodi.values())

# Построить график.
plt.plot(x_coords1, y_coords1, x_coords2, y_coords2)

# Показать график.
plt.show()
```
